```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
import pandas as pd
```

```python
train = pd.read_csv("iris_training.csv")
test = pd.read_csv("iris_test.csv")
```

```python
def input_fn(features, labels, training=True, batch_size=256):
    #convert the inputs to a dataset
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))
    
    #shuffle and repeat if yo are in training mode
    if training:
        dataset = dataset.shuffle(1000).repeat()
        
    return dataset.batch(batch_size)
```

```python
my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
```

# Building the model

```python
#defining the model in this case a deep neural network based model
#DNNClassifier is deep nerural network based classifier

classifier = tf.estimator.DNNClassifier(
    feature_columns = my_feature_columns,
    #two hidden layers for neural network of 30 and 10  respectively
    hidden_units = [30,10],
    #classes = 3, the model must chose between three classes
    n_classes = 3
)
```

# Training the model

```python
classifier.train(
    input_fn = lambda:input_fn(train, train_y, training=True),
    steps = 50000
)
input_fn
```

# Evaluating the model accuracy

```python
eval_result = classifier.evaluate(input_fn = lambda: input_fn(test, test_y, training = False))
print('\n Test set accuracy is: ', eval_result['accuracy'])
```

# Custom prediction on model

```python
def input_fn(features, batch_size=256):
    #converting the input datas to a dataset without labels for prediction
    return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

features = ['SepalLength', 'Sepalwidth', "PetalLength", "PetalWidth"]
predict_data = {}

print("Please enter the numeric datas to predict the class")
for feature in features:
    val= input("enter " + feature +":")
    predict_data[feature] = [float(val)]
    
predictions = classifier.predict(input_fn=lambda: input_fn(predict_data))
```
